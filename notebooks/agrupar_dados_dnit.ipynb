## Corrigir Bases de Dados

### Corrigir Base Acidentes

```python
def dividir_variaveis(coluna_classe):
    lista_classes = coluna_classe.to_list()
    tipo_pista = []
    tipo_area = []
    tipo_terreno = []
    for valor in lista_classes:
        # Identificar tipo pista
        if (type(valor)==float) and (math.isnan(valor)):
            tipo_pista.append('N/A')
        elif valor[0] == 'S':
            tipo_pista.append('Simples')
        elif valor[0] == 'D':
            tipo_pista.append('Dupla')
        else:
            tipo_pista.append('N/A')
        # Identificar tipo area
        if (type(valor)==float) and (math.isnan(valor)):
            tipo_area.append('N/A')
        elif valor[1] == 'R':
            tipo_area.append('Rural')
        elif valor[1] == 'U':
            tipo_area.append('Urbana')
        else:
            tipo_area.append('N/A')
        # Identificar tipo terreno
        if (type(valor)==float) and (math.isnan(valor)):
            tipo_terreno.append('N/A')
        elif valor[2] == 'P':
            tipo_terreno.append('Plano')
        elif valor[2] == 'O':
            tipo_terreno.append('Ondulado')
        elif valor[2] == 'M':
            tipo_terreno.append('Montanhoso')
        else:
            tipo_terreno.append('N/A')

    return pd.Series(tipo_pista, name='tipo_pista'), pd.Series(tipo_area, name='tipo_area'), pd.Series(tipo_terreno, name='tipo_terreno')
```

```python
tipo_pista, tipo_area, tipo_terreno = dividir_variaveis(acidentes['CLASSE'])
```

```python
acidentes['tipo_pista'] = tipo_pista
acidentes['tipo_area'] = tipo_area
acidentes['tipo_terreno'] = tipo_terreno
```

```python
acidentes.head()
```

```python
acidentes['SNV'].nunique()
```

```python
modas = acidentes.groupby('SNV')[['tipo_pista','tipo_area','tipo_terreno']].agg(
    lambda x: x.mode().iloc[0] if not x.mode().empty else None
)

somas_medidas = acidentes.groupby('SNV').agg(
    acidentes_sem_vitima=('acidentes_sem_vitima', 'sum'),
    acidentes_com_vitima=('acidentes_com_vitima', 'sum'),
    acidentes_com_mortos=('acidentes_com_mortos', 'sum'),
    acidentes_totais=('acidentes_totais', 'sum'),
    UPS_medio=('UPS', 'mean'),
    UPS_max=('UPS', 'max'),
    UPS_min=('UPS', 'min'),
    VDM=('VDM', 'mean'),
    ICA_medio=('ICA', 'mean'),
    ICA_max=('ICA', 'max'),
    ICA_min=('ICA', 'min'),
)

acidentes_trechos = somas_medidas.join(modas).reset_index()
```

```python
modas = contagem_trafego.groupby(['vl_codigo', 'vl_km_inic', 'vl_km_fina'])[['ds_sup_fed', 'GEH']].agg(
    lambda x: x.mode().iloc[0] if not x.mode().empty else None
)

somas_medidas = contagem_trafego.groupby(['vl_codigo', 'vl_km_inic', 'vl_km_fina']).agg(
    vl_extensa=('vl_ext_estim', 'sum'),
    VMDa_C=('VMDa_C', 'mean'),
    VMDa_D=('VMDa_D', 'mean')
)

trafego_trechos = somas_medidas.join(modas).reset_index()
```
